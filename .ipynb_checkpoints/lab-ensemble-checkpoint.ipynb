```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split

# Separate features and target
X = df.drop(columns=['Transported'])
y = df['Transported']

# Identify categorical and numerical features
categorical_features = X.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Preprocessing pipelines
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_pipeline, numerical_features),
    ('cat', categorical_pipeline, categorical_features)
])

# Final pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('feature_selection', SelectKBest(score_func=f_regression, k=10))
])

# Fit pipeline
pipeline.fit(X, y)

# Get feature names after one-hot encoding
encoded_feature_names = (
    pipeline.named_steps['preprocessing']
    .named_transformers_['cat']
    .named_steps['encoder']
    .get_feature_names_out(categorical_features)
)

all_feature_names = numerical_features + list(encoded_feature_names)

# Extract feature scores
selector = pipeline.named_steps['feature_selection']
scores = selector.scores_

# Display top features
feature_scores = pd.DataFrame({'Feature': all_feature_names, 'Score': scores})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)
print(feature_scores.head(10))
```

**Perform Train Test Split**

```python
#y = target
#X = features

y = df['FoodCourt']
X = df.drop(columns=['FoodCourt'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
X_train.head()
```

**Model Selection** - now you will try to apply different ensemble methods in order to get a better model

- Bagging and Pasting

```python
normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

```python
bagging_reg = BaggingRegressor(DecisionTreeRegressor(max_depth=20),
                               n_estimators=100,
                               max_samples = 1000)
```

```python
bagging_reg.fit(X_train_norm, y_train)
```

```python
#Evaluate model's performance

pred = bagging_reg.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", bagging_reg.score(X_test_norm, y_test))
```

- Random Forests

```python
#Initiaze random forests

forest = RandomForestRegressor(n_estimators=100, max_depth=20)
```

```python
#Training the model

forest.fit(X_train_norm, y_train)
```

```python
#Evaluate the model

pred = forest.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", forest.score(X_test_norm, y_test))
```

```python
#Bagging and Pasting : R2 score 0.46257023640369077
#Random Forest : R2 score 0.3845414026754267
```

- Gradient Boosting

```python
gb_reg = GradientBoostingRegressor(max_depth=20,
                                   n_estimators=100)
```

```python
#Training the model

gb_reg.fit(X_train_norm, y_train)
```

```python
#Evaluate the model

pred = gb_reg.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", gb_reg.score(X_test_norm, y_test))
```

```python
#Bagging and Pasting : R2 score 0.46257023640369077
#Random Forest : R2 score 0.3845414026754267
#Gradient Boosting : R2 score 0.1447679311700676
```

- Adaptive Boosting

```python
ada_reg = AdaBoostRegressor(DecisionTreeRegressor(max_depth=20),
                            n_estimators=100)
```

```python
#Training the model

ada_reg.fit(X_train_norm, y_train)
```

```python
#Evaluate the model

pred = ada_reg.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", ada_reg.score(X_test_norm, y_test))
```
